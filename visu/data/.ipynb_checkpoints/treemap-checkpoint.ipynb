```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
df = pd.read_csv('insertion_lp.csv')
```

```python
df = df.iloc[0:-2]
```

```python
list = []
data = {}
data['list'] = []

for i in range(len(df)):
    data['list'].append({
        'key': 'Insertion',
        'region': 'Master',
        'subregion': df.iloc[i]["discipline"],
        'value': str(int(df.iloc[i]["age_moyen"]))
    })
    data['list'].append({
        'key': 'En emploi',
        'region': 'Master',
        'subregion': df.iloc[i]["discipline"],
        'value': str(int(df.iloc[i]["en_emploi"]))
    })
    data['list'].append({
        'key': 'Emploi dans la région',
        'region': 'Master',
        'subregion': df.iloc[i]["discipline"],
        'value': str(int(df.iloc[i]["emploi_region"]))
    })
    data['list'].append({
        'key': 'Adéquation entre étude et emploi',
        'region': 'Master',
        'subregion': df.iloc[i]["discipline"],
        'value': str(int(df.iloc[i]["note"]))
    })
    data['list'].append({
        'key': 'Revenu',
        'region': 'Master',
        'subregion': df.iloc[i]["discipline"],
        'value': str(int(df.iloc[i]["revenu"]))
    })
    
with open('data.json', 'w') as outfile:  
    json.dump(data, outfile)
```
